```python
client = pymongo.MongoClient()
database = client['projet']
collection = database['Race']

#race = df_race.to_json(orient="records")
#race = json.loads(race)
#collection.insert_many(race)
```

```python
#cur = collection.find({"Grand_Prix":"Bahrain","Year":"2021"}).sort([("Points",-1)])
#list(cur)


# Best Speed average in the 2021 Bahrain Grand Prix


cur = collection.aggregate(
    [
        {"$match":{"Year":"2021","Grand_Prix":"Bahrain"}},
        {"$group": {"_id":"$Driver","Speed":{"$avg":"$Avg_Speed"}}},
        {"$sort":{"Speed":-1}},
        {"$limit":20}
    ]
)
list(cur)
```

```python
# Liste des constructeurs que le pilote a couru avec

cur = collection.aggregate(
    [
        {"$match":{"Driver":"Kvyat"}},
        {"$group": {"_id":"$Team"}},
        {"$limit":20}
    ]
)
list(cur)
```

```python
#Classement gÃ©nÃ©ral du pilote chaque annÃ©e

cur = collection.aggregate(
    [
        {"$match":{"Driver":"Vettel"}},
        {"$group": {"_id":"$Year","Position Averaged":{"$avg":"$Position"}}},
        {"$sort":{"Position Averaged":1}},
        #{"$limit":20}
    ]
)
list(cur)
```

```python
cur = collection.aggregate(
    [
        {"$match":{"Year":"2021"}},
        {"$group": {"_id":"$Team","Pitstop Time":{"$avg":"$Time_Pit"}}},
        {"$sort":{"Pitstop Time":1}},
        #{"$limit":20}
    ]
)
list(cur)
```

```python
cur = collection.aggregate(
    [
        {"$match":{"Driver":"Verstappen"}},
        {"$group": {"_id":"$Grand_Prix","Pitstop Time":{"$avg":"$Time_Pit"}}},
        {"$sort":{"Pitstop Time":1}},
        #{"$limit":20}
    ]
)
list(cur)
```

```python
cur = collection.aggregate(
    [
        {"$match":{"Year":"2021",'Driver':"Verstappen"}},
        {"$project": {"_id":{"Grand Prix":"$Grand_Prix"},
                        "Time":{
                            "$dateFromString":{
                                "dateString":"$Q3",
                                "format":"%M:%S.%LZ",
                                "onError": '$Q3',
                                "onNull":"DNF"}
        }}},
        #{"$group": {"_id":"$Team"}},
        {"$sort":{"Time":1}},
        #{"$limit":20}
    ]
)
list(cur)
```

```python
#Nombre de points du constructeur chaque annÃ©e

cur = collection.aggregate(
    [
        {"$match":{"Team":"Mercedes"}},
        {"$group": {"_id":"$Year","Total Points":{"$sum":"$Points"}}},
        {"$sort":{"Total Points":1}},
        #{"$limit":20}
    ]
)
list(cur)
```

```python
#Nombre de victoire du pilote chaque annÃ©e


cur = collection.aggregate(
    [
        {"$match":{"Driver":"Hamilton","Position":1}},
        {"$group": {"_id":"$Year","Number of wins":{"$sum":1}}},
        {"$sort":{"Number of wins":1}},
        #{"$limit":20}
    ]
)
list(cur)
```

```python
# Nombre d'abandon par annÃ©e
cur = collection.aggregate(
    [
        {"$match":{"Gap_Time":"DNF"}},
        {"$group": {"_id":"$Year","Number of DNF":{"$sum":1}}},
        {"$sort":{"Number of DNF":1}},
        #{"$limit":20}
    ]
)
list(cur)
```

```python
# Nombre d'abandon par annÃ©e sur un Grand Prix
cur = collection.aggregate(
    [
        {"$match":{"Gap_Time":"DNF","Grand_Prix":"Monaco"}},
        {"$group": {"_id":"$Year","Number of DNF":{"$sum":1}}},
        {"$sort":{"Number of DNF":1}},
        #{"$limit":20}
    ]
)
list(cur)
```

```python
# Grand Prix avec le plus de DNF dans la decennie
cur = collection.aggregate(
    [
        {"$match":{"Gap_Time":"DNF"}},
        {"$group": {"_id":"$Grand_Prix","Number of DNF":{"$sum":1}}},
        {"$sort":{"Number of DNF":-1}},
        #{"$limit":20}
    ]
)
list(cur)
```

```python
# Driver and Number of DNF in each Grand Prix
cur = collection.aggregate(
    [
        {"$match":{"Gap_Time":"DNF"}},
        {"$group": {"_id":{"Grand Prix":"$Grand_Prix","Driver":"$Driver"},"Number of DNF":{"$sum":1}}},
        {"$sort":{"Number of DNF":-1}},
        {"$limit":20}
    ]
)
list(cur)
```

```python
# Driver with most pole position in the decade
cur = collection.aggregate(
    [
        {"$match":{"Pos_Qualif":"1"}},
        {"$group": {"_id":"$Driver","Number of Pole Position":{"$sum":1}}},
        {"$sort":{"Number of Pole Position":-1}},
        {"$limit":20}
    ]
)
list(cur)
```

```python
# Driver with most win in the decade
cur = collection.aggregate(
    [
        {"$match":{"Position":1}},
        {"$group": {"_id":"$Driver","Number of Wins":{"$sum":1}}},
        {"$sort":{"Number of Wins":-1}},
        {"$limit":20}
    ]
)
list(cur)
```

```python
# Averaged Max Speed for each Grand Prix

cur = collection.aggregate(
    [
        {"$group": {"_id":"$Grand_Prix","Speed Averaged":{"$avg":"$Avg_Speed"}}},
        {"$sort":{"Speed Averaged":-1}},
        #{"$limit":20}
    ]
)
list(cur)
```

```python
# Rank of fastest pilots in a team during the decade

cur = collection.aggregate(
    [
        {"$match":{"Team":"McLaren Mercedes"}},
        {"$group": {"_id":"$Driver","Speed Averaged":{"$avg":"$Avg_Speed"}}},
        {"$sort":{"Speed Averaged":-1}},
        #{"$limit":20}
    ]
)
list(cur)
#McLaren Mercedes
```
